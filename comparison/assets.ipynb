## Extracting data

```python
import pandas as pd
import numpy as np
import matplotlib

%matplotlib inline

matplotlib.rc('figure', figsize=(20,10))
matplotlib.rc('font', weight='bold', size=12)
```

```python
d2004 = pd.read_csv('2004.csv')
d2009 = pd.read_csv('2009.csv')
d2014 = pd.read_csv('2014.csv')
d2019 = pd.read_csv('2019.csv')
```

```python
ad2004 = d2004[["Candidate","Constituency","Party", "Total Assets"]].sort_values('Total Assets', ascending=False).head(1000)
ad2004 = ad2004.reset_index(drop=True)
ad2004.groupby("Party").aggregate(np.sum).sort_values("Total Assets",ascending=False).plot(kind='bar')
```

```python
ad2009 = d2009[["Candidate","Constituency","Party", "Total Assets"]].sort_values('Total Assets', ascending=False).head(1000)
ad2009 = ad2009.reset_index(drop=True)
ad2009.groupby("Party").aggregate(np.sum).sort_values("Total Assets",ascending=False).plot(kind='bar')
```

```python
ad2014 = d2014[["Candidate","Constituency","Party", "Total Assets"]].sort_values('Total Assets', ascending=False).head(1000)
ad2014 = ad2014.reset_index(drop=True)
ad2014.groupby("Party").aggregate(np.sum).sort_values("Total Assets",ascending=False).plot(kind='bar')
```

```python
ad2019 = d2019[["Candidate","Constituency","Party", "Total Assets"]].sort_values('Total Assets', ascending=False).head(1000)
ad2019 = ad2019.reset_index(drop=True)
ad2019.groupby("Party").aggregate(np.sum).sort_values("Total Assets",ascending=False).plot(kind='bar')
```
